## 1.1 Introduction to Object-Oriented Programming (OOP)

### 1. What is OOP?
Imagine you are building a video game with thousands of characters. Instead of writing separate code for every single character, you create a **"Blueprint"** that defines what every character has (like a name and health) and what every character can do (like walk or attack).

- **Class**: The "Blueprint" or "Template" (e.g., the concept of a "Dog").
- **Object (Instance)**: The actual "Thing" created from the blueprint (e.g., your pet dog "Buddy").

### 2. Core Keywords
In Python, we use these key terms:
- `class`: The keyword used to define a new blueprint.
- `__init__`: A special method called the "Constructor." It runs automatically when you create a new object to set its initial data.
- `self`: A reference to the specific object being created. It allows the object to "remember" its own data (e.g., `self.name`).
- **Method**: A function defined inside a class that represents an action the object can perform.

```python
# 1. Define the Blueprint (Class)
class Dog:
    def __init__(self, name, breed):
        # Attributes: Data stored in the object
        self.name = name
        self.breed = breed

    # Method: Behavior of the object
    def bark(self):
        print(f"Woof! My name is {self.name} and I am a {self.breed}.")

# 2. Create actual Objects (Instances) from the Blueprint
dog1 = Dog("Buddy", "Golden Retriever")
dog2 = Dog("Max", "Beagle")

# 3. Accessing attributes and calling methods
print(f"First dog's name: {dog1.name}")
dog1.bark()

print(f"Second dog's name: {dog2.name}")
dog2.bark()
```

## 1.2 Case Study — The Geometry of a Triangle

To see the real power of OOP, let's look at a geometric Triangle.

A triangle is defined by its three sides ($a, b, c$). From just these three numbers, we can calculate:
1. **Area** (using Heron's Formula)
2. **Heights** for each side
3. **Incircle Area** (the circle inside)
4. **Circumcircle Area** (the circle outside)

### The Comparison:
- **Functional Approach**: We store side lengths in loose variables and pass them into many different functions. If we have 10 triangles, we have 30 variables to track!
- **OOP Approach**: We create a `Triangle` object. It "knows" its own sides and "knows" how to calculate its own properties. It is a self-contained, intelligent package.

### Funtional Approach

```python
import math

# Data is stored as loose variables
side_a, side_b, side_c = 3, 4, 5

def get_area(a, b, c):
    s = (a + b + c) / 2
    return math.sqrt(s * (s - a) * (s - b) * (s - c))

def get_heights(a, b, c):
    area = get_area(a, b, c)
    return (2*area/a, 2*area/b, 2*area/c)

def scale_triangle(a, b, c, n):
    # We have to return a tuple and update variables manually
    return a * n, b * n, c * n

# Usage is cumbersome:
print(f"Area: {get_area(side_a, side_b, side_c)}")
print(f"Heights: {get_heights(side_a, side_b, side_c)}")

# Scaling requires re-assigning variables
side_a, side_b, side_c = scale_triangle(side_a, side_b, side_c, 2)
print(f"New Area after scaling: {get_area(side_a, side_b, side_c)}")

# Imagine if you had 5 different triangles... it would be a nightmare to track!
```

### OOP Approach

```python
import math

class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    @property
    def area(self):
        """Calculates area automatically using Heron's formula"""
        s = (self.a + self.b + self.c) / 2
        return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))

    @property
    def heights(self):
        """Returns the heights for sides a, b, and c"""
        return (2 * self.area / self.a, 2 * self.area / self.b, 2 * self.area / self.c)

    @property
    def incircle_area(self):
        """Area of the inscribed circle: r = Area / semi-perimeter"""
        s = (self.a + self.b + self.c) / 2
        r = self.area / s
        return math.pi * (r ** 2)

    @property
    def circumcircle_area(self):
        """Area of the circumscribed circle: R = (abc) / (4 * Area)"""
        R = (self.a * self.b * self.c) / (4 * self.area)
        return math.pi * (R ** 2)

    def scale(self, n):
        """Resizes the triangle by factor n"""
        self.a *= n
        self.b *= n
        self.c *= n
        print(f"Triangle scaled by {n}x")

# --- Elegant Usage ---

# Create a triangle object
tri = Triangle(3, 4, 5)

# Access properties like they are simple data (no parentheses needed thanks to @property)
print(f"Initial Area: {tri.area:.2f}")
print(f"Heights: {tri.heights}")
print(f"Incircle Area: {tri.incircle_area:.2f}")

# Scale it!
tri.scale(2)

# The area and heights update AUTOMATICALLY because they depend on the updated sides
print(f"New Area: {tri.area:.2f}")
print(f"New Circumcircle Area: {tri.circumcircle_area:.2f}")
```

## OOTD Advisor

I believe all of you may check the weather forecast before you decide OOTD(Outfit of The Day). But do you know Hong Kong Observation also offers forecast API for public? So we can construct an OOTD advisor by:

1. Call HKO API to fetch the forecast information.
2. Send the forecast to AI via API to advice OOTD.

```python
import requests
from datetime import datetime, timedelta
from openai import OpenAI

# ---------- Step 0: Configure LLM client (Method 1: pass key directly) ----------
client = OpenAI(
    api_key="sk-5f4PbxaWl87YovCPfONeUl8lzyhW4qPkKkR3a33RJ7fDocd5",
    base_url="https://api.bianxie.ai/v1"  # Bianxie AI aggregation API
)

# ---------- Step 1: Call Hong Kong Observatory (HKO) API for 9‑day forecast ----------
hko_url = "https://data.weather.gov.hk/weatherAPI/opendata/weather.php"
params = {
    "dataType": "fnd",  # 9‑day forecast
    "lang": "tc"        # Traditional Chinese descriptions
}
hko_resp = requests.get(hko_url, params=params)
hko_resp.raise_for_status()

forecast_data = hko_resp.json()

# Pick tomorrow's forecast
today = datetime.now().date()
tomorrow = today + timedelta(days=1)
tomorrow_str = tomorrow.strftime("%Y%m%d")

weather_forecast_list = forecast_data.get("weatherForecast", [])
tomorrow_forecast = None
for item in weather_forecast_list:
    if item.get("forecastDate") == tomorrow_str:
        tomorrow_forecast = item
        break

if tomorrow_forecast is None:
    raise ValueError("Cannot find weather forecast for tomorrow in HKO API response.")

# Format a short summary of tomorrow's weather
tomorrow_summary = (
    f"Date: {tomorrow_forecast.get('week')} ({tomorrow_forecast.get('forecastDate')})\n"
    f"Weather: {tomorrow_forecast.get('forecastWeather')}\n"
    f"Wind: {tomorrow_forecast.get('forecastWind')}\n"
    f"Temperature: "
    f"{tomorrow_forecast['forecastMintemp']['value']}–"
    f"{tomorrow_forecast['forecastMaxtemp']['value']} "
    f"{tomorrow_forecast['forecastMaxtemp']['unit']}\n"
    f"Relative humidity: "
    f"{tomorrow_forecast['forecastMinrh']['value']}–"
    f"{tomorrow_forecast['forecastMaxrh']['value']}%"
)

print("=== HKO forecast for tomorrow ===\n")
print(tomorrow_summary)

# ---------- Step 2: Send the forecast to LLM as an OOTD advisor ----------
question_for_llm = f"""
You are an OOTD (Outfit of The Day) advisor for a person living in Hong Kong.

Here is the weather forecast for tomorrow from the Hong Kong Observatory (HKO):
{tomorrow_summary}

Based on this forecast:
1. Briefly summarize the expected weather in English.
2. Recommend a suitable outfit for tomorrow (tops, bottoms, shoes, and optional accessories).
3. Explain in 2–3 sentences why this outfit is appropriate for the weather.
"""

completion = client.chat.completions.create(
    model="gpt-4-all",
    messages=[
        {
            "role": "user",
            "content": question_for_llm,
        }
    ]
)

print("\n=== OOTD Advisor (LLM answer) ===\n")
print(completion.choices[0].message.content)
```
